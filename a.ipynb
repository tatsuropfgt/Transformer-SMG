```python
import torch

p = torch.tensor([55, 79, 76, 48, 77])
p_pt = p.view(5, 1) - p.view(1, 5)
upper_triangle = torch.triu(torch.ones(5, 5), diagonal=1)
p_pt[upper_triangle == 1] = -40
```

```python
p_pt
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

viridis = plt.cm.get_cmap('viridis', 256)

newcolors = viridis(np.linspace(0, 1, 256))
black = np.array([0, 0, 0, 1])
newcolors[0] = black
newcmp = ListedColormap(newcolors)


fig, ax = plt.subplots()
cax = ax.imshow(p_pt, cmap=newcmp, interpolation='none')
ax.set_xticks([])
ax.set_yticks([])

for i in range(p_pt.shape[0]):
    for j in range(p_pt.shape[1]):
        if p_pt[i, j] != -40:
            if p_pt[i, j] > 0:
                text = f"+{p_pt[i,j]}"
            elif p_pt[i, j] < 0:
                text = f"-{-p_pt[i,j]}"
            else:
                text = "±0"

            ax.text(j, i, text, ha='center', va='center', color='white', fontsize=13)

plt.show()
```

```python
p_pt_oct = p_pt // 12
p_pt_semi = p_pt % 12
p_pt_semi[p_pt_oct == -4] = -1
p_pt_semi
```

```python
fig, ax = plt.subplots()
cax = ax.imshow(p_pt_oct, cmap=newcmp, interpolation='none')
ax.set_xticks([])
ax.set_yticks([])

for i in range(p_pt_oct.shape[0]):
    for j in range(p_pt_oct.shape[1]):
        if p_pt_oct[i, j] != -4:
            if p_pt_oct[i, j] > 0:
                text = f"+{p_pt_oct[i,j]}"
            elif p_pt_oct[i, j] < 0:
                text = f"-{-p_pt_oct[i,j]}"
            else:
                text = "±0"

            ax.text(j, i, text, ha='center', va='center', color='white', fontsize=13)

plt.show()
```

```python
fig, ax = plt.subplots()
cax = ax.imshow(p_pt_semi, cmap=newcmp, interpolation='none')
ax.set_xticks([])
ax.set_yticks([])

for i in range(p_pt_semi.shape[0]):
    for j in range(p_pt_semi.shape[1]):
        if p_pt_semi[i, j] != -1:
            if p_pt_semi[i, j] > 0:
                text = f"+{p_pt_semi[i,j]}"
            elif p_pt_oct[i, j] < 0:
                text = f"-{-p_pt_semi[i,j]}"
            else:
                text = "±0"

            ax.text(j, i, text, ha='center', va='center', color='white', fontsize=13)

plt.show()
```
